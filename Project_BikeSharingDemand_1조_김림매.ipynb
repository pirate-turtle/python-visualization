```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data['datetime'] = data['datetime'].astype('datetime64')
```

```python
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
data['day'] = data['datetime'].dt.day
data['hour'] = data['datetime'].dt.hour
data['minute'] = data['datetime'].dt.minute
data['second'] = data['datetime'].dt.second
```

```python
# 가설: 가을에 저전거를 더 많이 대여한다.
```

```python
sns.barplot(data=data, x="season", y="count", hue="holiday")
```

```python
# 가설: 습도가 높을 수록 자전거를 적게 대여한다.
```

```python
plt.figure(figsize=(16, 5))
sns.lineplot(data=data, x="humidity", y="count")
```

```python
# 가설: 날씨가 따뜻할 수록 자전거를 많이 대여한다.
```

```python
plt.figure(figsize=(16, 5))
sns.lineplot(data=data, x="temp", y="count",  hue='weather')
```

```python
#가설: 가을 때 체감온도가 적절하고 자전거을 더 많이 대여한다. 
```

```python
plt.figure(figsize=(16, 9))
sns.pointplot(data=data, x="atemp", y="count", hue='weather')
```
