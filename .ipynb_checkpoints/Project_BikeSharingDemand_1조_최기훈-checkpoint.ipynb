```python
## 가설 

## 1. 온도가 높으면 대여량이 감소할 것이다
## 2. 회원과 비회원의 온도별 대여량 추이는 같을 것이다.
## 3. 계졀별 대여량 차이가 날 것이다
## 4. 해질녘 - 저녁에 자전거를 많이 탈 것이다. 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data['datetime'] = data['datetime'].astype('datetime64')
data.dtypes
```

```python
data['year'] =data['datetime'].dt.year
data['month'] =data['datetime'].dt.month
data['day'] =data['datetime'].dt.day
data['hour'] =data['datetime'].dt.hour
data['minute'] =data['datetime'].dt.minute
data['second'] =data['datetime'].dt.second
```

```python
## 회원과 비회원의 온도별 대여량 추이는 같을 것이다.
f, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

sns.lineplot(data = data, x='temp', y='registered', palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=True)
ax1.set_ylabel("registered")

sns.lineplot(data = data, x='temp', y='casual', palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=True)
ax2.set_ylabel("casual")
plt.title 
```

```python
## 낮은 습도와 적당한 온도에서 대여량이 늘 것이다.
plt.figure(figsize=(12,6))
sns.scatterplot(data=data,x='temp',y='humidity',hue='count')
```

```python
## 계졀별 대여량 차이가 날 것이다
plt.figure(figsize=(12,6))
sns.boxplot(data=data,x='season',y='count')
```

```python
## 해질녘 - 저녁에 자전거를 많이 탈 것이다. 
plt.figure(figsize=(12,6))
sns.swarmplot(data=data,x="hour",y="count")
```
