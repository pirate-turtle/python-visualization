```python
import pandas as pd 
pd.plotting.register_matplotlib_converters() 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
print("Setup Complete") 
```

```python
# 읽어들일 파일 경로
flight_filepath = "./dataset/flight_delays.csv" 

# 변수 flight_data에 파일을 읽어들임 
flight_data = pd.read_csv(flight_filepath, index_col="Month")

# 데이터 출력 
flight_data
```

```python
sns.barplot(x=flight_data.index, y=flight_data['NK'], data=flight_data)
```

```python
# 도표의 너비(10)와 높이(6) 설정
plt.figure(figsize=(10,6)) 

# 도표의 제목 설정
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month") 

# 막대그래프 그리기
sns.barplot(x=flight_data.index, y=flight_data['NK']) 

# 세로축 레이블 생성
plt.ylabel("Arrival delay (in minutes)") 
```

# 실습과제
## American Airlines (항공사코드: AA)의 월별 평균 비행 지연 시간

```python
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 8), sharex=True)

# ax1, ax2, ax3는 각각
sns.barplot(x=flight_data.index, y=flight_data['NK'], data=flight_data, palette="rocket", ax=ax1)
ax1.set_ylabel("NK")
ax1.set_xlabel("")

sns.barplot(x=flight_data.index, y=flight_data['AA'], data=flight_data, palette="RdBu", ax=ax2)
ax2.set_ylabel("AA")
ax2.set_xlabel("")

sns.barplot(x=flight_data.index, y=flight_data['F9'], data=flight_data, palette="Blues_d", ax=ax3)
ax3.set_ylabel("F9")
```
